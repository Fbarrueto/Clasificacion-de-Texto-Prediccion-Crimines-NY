```python
def words_count_dict(df):
    words = []
    for palabra in df['tweet_without_stopwords']: 
        words.extend(palabra)

    c = Counter(words)
    return c
```

```python
count_words_positive = words_count_dict(df_positive)
count_words_negative = words_count_dict(df_negative)
```

```python
type(count_words_positive)
```

```python
def positive_negative(tweet,sentiment,count_words_positive,count_words_negative):
    if sentiment == 'Neutral':
        count = 0
        for word in tweet:
            diff = count_words_positive[word] - count_words_negative[word]
            count += diff

        if count > 0:
            est_sentiment= 'Positiva'

        if count < 0:
            est_sentiment= 'Negativa'

        else:
            rand = random.choice(['Positiva', 'Negativa'])
            est_sentiment = rand

        return est_sentiment
    else:
        return sentiment
```

```python
df['sentiment'] = df.apply(lambda tweet : positive_negative(tweet['content'],tweet['sentiment'],count_words_positive,count_words_negative), axis=1)
```

```python
df.head()
```

Procederemos a lemmatizar las palabras de la columna tweets sin stopwords, con la finalidad de llegar a obtener la reducción del tronco lexico. Crearemos 2 funciones, la primera permitira aplicar el metodo WordNetLemmatizer de la librería NLTK a las palabras del corpus, y en la segunda uniremos estas palabras lematizadas para insertarlas dentro del dataset que trabajamos.

```python
def lemmatization(tweet):
    lemma = WordNetLemmatizer()
    normalized_sentence = []
    for word in tweet:
        normalized_sentence.append(lemma.lemmatize(word))
        
    return normalized_sentence
```

```python
df['tweet_without_stopwords'] = df.apply(lambda tweet: lemmatization(tweet['tweet_without_stopwords']), axis=1)
```

```python
def join_words(tweet):
    join_words = ' '.join([word for word in tweet])
    return join_words
```
